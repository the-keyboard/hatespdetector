```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
import re
import nltk
from nltk.util import pr
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword = set(stopwords.words("english"))
```

```python
df=pd.read_csv("twitter_data.csv")
print(df.head())
```

```python
df['labels']=df['class'].map({0: "Potential Hate Speech Detected", 1:"Potential Offensive Language Detetcted", 2: "Clean"})
print(df.head())
```

```python
df = df [['tweet', 'labels' ]]
df.head()
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub(r'\[.*?\]', '', text)  # Use raw string to avoid invalid escape sequence
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Use raw string to avoid invalid escape sequence
    text = re.sub(r'<.*?>+', '', text)  # Use raw string to avoid invalid escape sequence
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # Fixed syntax error
    text = re.sub(r'\n', '', text)  # Use raw string to avoid invalid escape sequence
    text = re.sub(r'\w*\d\w*', '', text)  # Use raw string to avoid invalid escape sequence
    text = [word for word in text.split() if word not in stopword]  # Fixed split argument
    text = " ".join(text)
    return text

# Assuming df is your DataFrame and "tweet" is the column you want to clean
df["tweet"] = df["tweet"].apply(clean)
print(df.head())
```

```python
x = np.array(df["tweet"])
y = np.array(df["labels"])

# Vectorize the text data
cv = CountVectorizer()
x = cv.fit_transform(x)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# Train a Decision Tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
test_data="served cunt fr"
df=cv.transform([test_data])
print(clf.predict(df))
```
